# Rock Leachate Analysis (Proxy)

This notebook explores USGS NWIS instantaneous discharge data used as a proxy
for leachate flow dynamics across monitoring stations.


```python
import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH = "../data/raw/leachate.csv"

df = pd.read_csv(DATA_PATH)
print(df.shape)
df.head()
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

summary = {
    "num_rows": len(df),
    "num_rocks": df["rock_id"].nunique(),
    "start": df["timestamp"].min(),
    "end": df["timestamp"].max(),
}
summary
```

```python
plt.figure(figsize=(6, 4))
df["target"].hist(bins=50)
plt.title("Target Distribution (Discharge)")
plt.xlabel("Discharge")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
example_rock = df["rock_id"].iloc[0]
rock_df = df[df["rock_id"] == example_rock].sort_values("timestamp")

plt.figure(figsize=(8, 3))
plt.plot(rock_df["timestamp"], rock_df["target"])
plt.title(f"Discharge Over Time: {example_rock}")
plt.xlabel("Time")
plt.ylabel("Discharge")
plt.tight_layout()
plt.show()
```
